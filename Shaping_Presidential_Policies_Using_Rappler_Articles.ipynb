# **Methodology**
## Data Collection
Our data collection process focused on gathering information from Rappler, specifically targeting content related to the query `marcos year 2022`. The first ten pages of search results were downloaded as HTML files. From these files, we extracted critical information, organizing the titles, links, publication dates, and snippets of each article into a structured table. To deepen our dataset, we then accessed each link individually to retrieve the full text of the articles. This effort resulted in a comprehensive dataset, where each row represented the complete text of a single article.

## **Data Processing**
In the data processing phase, we applied the Bag of Words model, a method crucial for converting text into numerical features, which is essential for computational models to understand and process language. This model treats each word in the document as an individual feature, disregarding grammar and word order, but maintaining frequency, which helps in understanding the prominence of certain terms and themes.

Following this, we implemented TF-IDF (Term Frequency-Inverse Document Frequency), an approach particularly effective for text-heavy datasets. TF-IDF helps in distinguishing the importance of words in a document relative to a collection of documents, thus enhancing the Bag of Words model by reducing the weight of commonly used words and increasing the significance of unique terms. Finally, we applied Latent Semantic Analysis (LSA), a technique that aids in identifying patterns and relationships between terms and concepts in large text datasets. LSA was utilized as a key technique for determining the dominant themes within our dataset.

The singular vectors derived from LSA were subjected to K-Means clustering. This step was crucial for categorizing the identified themes into coherent groups, enhancing the clarity and specificity of our findings. To ensure the effectiveness of our clustering approach, we employed both the silhouette method and the Calinski-Harabasz index. The silhouette method provided a measure of how similar an object is to its own cluster compared to other clusters, offering insights into the appropriateness of the number of clusters. Meanwhile, the Calinski-Harabasz index helped in assessing the cluster validity by comparing the dispersion between and within clusters. These metrics were instrumental in fine-tuning our clustering approach, ensuring that the K-Means algorithm optimally grouped the themes identified by LSA.

## **Web Scraping Proper**
The first 10 pages of search results from Rappler were previously downloaded. Webpages in HTML format are then uploaded in a folder in Jojie as a central repository. This allows us to have a static source of generating the steps for this study and to support consistent replication in the future.

```python
# Set the file directory
data_dir = '/mnt/processed/private/msds2024/lt3'
```

```python
# Create a DataFrame from the Rappler html files
articles = []
for i in range(1, 11):
    with open(os.path.join(data_dir, f"Search with RAPPLER{i}.html")) as f:
        soup = BeautifulSoup(f).find_all(
            "div", class_="gs-bidi-start-align gs-snippet", dir="ltr"
        )
    for item in soup:
        articles.append(
            {
                "Title": item.find_previous("a").text.strip(),
                "Link": item.find_previous("a")["href"],
            }
        )
df = pd.DataFrame(articles)
```

# **Data Processing**
This section demonstrates the representation of documents into Bag-of-Words, information retrieval, and performance evaluation.

## **Bag-of-Words Representation**
Documents are represented as bag-of-words based on the defined regular expression (Regex) capturing any string with three or more alphanumeric characters `\b[a-z]{3,}\b`.

```python
# Create the corpus from the Rappler links
corpus = []
for link in df["Link"]:
    corpus.append(
        BeautifulSoup(requests.get(link).text)
        .find("div", class_="post-single__content entry-content")
        .text
    )
    time.sleep(0.2)
```

```python
print('Length of Corpus: ', len(corpus))
```

## **TF-IDF**
The bag-of-words were subjected to TF-IDF with the same token pattern `\b[a-z]{3,}\b` and `stop_words="english"`

We set `min_df=5` to select words that appear at least 5 times and `max_df=0.8` for words that appear at most 80% of the documents.

```python
# Apply the TfidfVectorizer to create a bag-of-words with TF-IDF
tfidf = TfidfVectorizer(
    min_df=5, max_df=0.8, token_pattern=r"\b[a-z]{3,}\b", stop_words="english"
)
bow = tfidf.fit_transform(corpus)
df_bow = pd.DataFrame(bow.toarray(), columns=tfidf.get_feature_names_out())
```

```python
display(df_bow)
display(HTML('<center><i>Table 2. Preview of Bag-of-Words</i></center>'))
```

## **Determining the Number of Singular Vectors**
Using the plot of `Variance Explained` vs `SV`, we determine how much Singular Vectors we have to retain, analyze, and cluster, with the cumulative variance of 0.9.

```python
# Plot the cumulative variance explained and find the number of SVs to retain
pct = 0.9
svd = TruncatedSVD(n_components=max(bow.shape))
svd.fit(bow)
var = svd.explained_variance_ratio_
n = np.where(var.cumsum() > pct)[0][0] + 1
fig, ax = plt.subplots()
ax.plot(range(1, len(var)+1), var, '-', label='individual')
ax.plot(range(1, len(var) + 1), var.cumsum(), label='cumulative')
ax.axhline(pct, ls="--", color="g")
ax.axvline(n, ls="--", color="g")
plt.xlabel("SV")
plt.ylabel('Variance Explained')
ax = ax.twinx()
ax.set_ylabel("Cumulative Variance Explained")
fig.text(
    0.5,
    -0.1,
    f"Fig 2. Variance explained vs SVs retained\n"
    f"If we want to retain a cumulative variance explained of {pct},\n"
    f"then we have to retain {n} singular vectors.",
    wrap=True,
    horizontalalignment="center",
    fontsize=12,
)
plt.show()
```

## **Truncated SVD and Exploring First Five SVs**
In order to highlight the most significant singular values in the data, we subjected the bag-of-words to TruncatedSVD, and we analyze the first five SVs:


*   *SV1: Unity Tandem.*
*   *SV2: Foreign Policy*
*   *SV3: Fake News*
*   *SV4: Commodity Prices*
*   *SV5: COMELEC Case*

and so on...


```python
# Apply TruncatedSVD to the bag-of-words
svd = TruncatedSVD(n_components=n)
bow_svd = svd.fit_transform(df_bow)
V_T = svd.components_.T
```

```python
feature_names = tfidf.get_feature_names_out()
fig, axs = plt.subplots(5, 1, figsize=(6, 20))
for i in range(5):
    # fig, ax = plt.subplots()
    order = np.argsort(np.abs(V_T[:, i]))[-10:]
    axs[i].barh([feature_names[o] for o in order], V_T[order, i])
    axs[i].set_title(f"SV{i+1}")
fig.text(
    0.5,
    0.08,
    "Fig 3. LSA of the first 5 SVs",
    wrap=True,
    horizontalalignment="center",
    fontsize=12,
)
plt.show()
```

## **K-Means Clustering with Silhouette Score and Calinksi Harabasz**
Because manually clustering the SVs would result to further bias, it would be logical to subject them to K-Means Clustering instead, with the Silhouette Score and Calinksi Harabasz to determine the ideal number of clusters. From the plots below, we determine that the ideal number of clusters is `k=3`.

```python
# Instantiate the clustering model and elbow visualizer
model = KMeans(n_init="auto", random_state=0)
visualizer = KElbowVisualizer(
    model, k=(2, 12), metric="silhouette", timings=False, title=" "
)

# Fit the data to the visualizer and render the figure
fig, ax = plt.subplots()
visualizer.fit(bow_svd)
fig.text(
    0.5,
    -0.02,
    "Fig 4. Silhouette Score Elbow for KMeans Clustering",
    wrap=True,
    horizontalalignment="center",
    fontsize=12,
)
visualizer.show(ax=ax);
```

```python
# Instantiate the clustering model and elbow visualizer
model = KMeans(n_init="auto", random_state=0)
visualizer = KElbowVisualizer(
    model, k=(2, 12), metric="calinski_harabasz", timings=False, title=" "
)

# Fit the data to the visualizer and render the figure
fig, ax = plt.subplots()
visualizer.fit(bow_svd)
fig.text(
    0.5,
    -0.02,
    "Fig 5. Calinski Harabasz Score Elbow for KMeans Clustering",
    wrap=True,
    horizontalalignment="center",
    fontsize=12,
)
visualizer.show(ax=ax);
```

## **Plotting the K-Means Cluster**
When the Bag of Words are visualized, the three clusters and their distinctions from one another is evident. From here, we are able to directly label each cluster from their words.

```python
# Fit the bag-of-words with KMeans
kmeans = KMeans(
    n_clusters=visualizer.elbow_value_, n_init="auto", random_state=0
)
kmeans.fit(bow_svd)

# Plot the KMeans clusters
fig, ax = plt.subplots()
ax.scatter(bow_svd[:, 0], bow_svd[:, 1], c=kmeans.labels_)
plt.xlabel("SV1")
plt.ylabel("SV2")
fig.text(
    0.5,
    -0.02,
    f"Fig 6. KMeans Clustering with n_clusters={visualizer.elbow_value_}",
    wrap=True,
    horizontalalignment="center",
    fontsize=12,
)
plt.show()
```

```python
# Find the most influential words per cluster
original_space_centroids = svd.inverse_transform(kmeans.cluster_centers_)
order_centroids = original_space_centroids.argsort()[:, ::-1]
terms = tfidf.get_feature_names_out()

print("Most Influential Words per Cluster")

for i in range(kmeans.cluster_centers_.shape[0]):
    print(f"\033[3{i+1}mCluster {i}: ", end="")
    for ind in order_centroids[i, :10]:
        print(f"{terms[ind]} ", end="")
    print()
```

```python
# Print the article titles by cluster
print("Article Titles by Cluster")
df_cluster = pd.DataFrame([df["Title"], kmeans.labels_]).T.rename(
    columns={"Unnamed 0": "Label"}
)

for label in df_cluster.Label.unique():
    for title in df_cluster.set_index("Label")["Title"][label]:
        print(f"\033[3{label+1}m{title}")
```

# **Analysis and Results**

After performing K-means clustering, we arrived at three dominant clusters.
* `Cluster 0`: duterte governor norte party sara bongbong presidential ilocos family vice
* `Cluster 1`: media disinformation accounts comelec journalists gold duterte petition twitter family
* `Cluster 2`: china singapore sona visit government state economic year philippine prices

We can call these clusters as follows:
* `Cluster 0`: **"Political Alliances and Dynasties"**

  This cluster seems to focus on political figures, regional governance, and connections in politics. The keywords like `duterte`, `bongbong`, `presidential`, `vice`, and `party` suggest a focus on political personalities and their influence in regional and national politics.
  While the exemplar of our database explicitly mentioned `marcos`, we can see a strong correlation with this word to other political personalities and affiliations. Ferdinand ‘Bongbong’ Marcos Jr. is paired with Sara Duterte and their alliance in the 2022 election has shown to be a formidable force after getting more than 50% of the popular votes for both presidential and vice presidential posts.

* `Cluster 1`: **"Media and Disinformation"**

  This cluster includes keywords such as `media`, `journalists`, and `twitter`, pointing to the role of media and social media in the political landscape. Meanwhile, keywords like `disinformation`, and `gold`, highlight a critical threat to the integrity of media, and journalism.

* `Cluster 2`: **"International Relations and Economic Policies"**

  Keywords like `china`, `singapore`, `sona` (State of the Nation Address), `economic`, and `government` suggest a focus on international relations, economic policies, and government strategies. This cluster likely covers the country's diplomatic engagements and economic initiatives, as well as their impact on local economic factors like prices.
